```python
raw_train_dataset.drop(columns=["id", "Unnamed: 0"], inplace=True)
raw_test_dataset.drop(columns=["id", "Unnamed: 0"], inplace=True)
```

```python
raw_train_dataset["Arrival Delay in Minutes"].fillna(raw_train_dataset["Departure Delay in Minutes"], inplace=True)
raw_test_dataset["Arrival Delay in Minutes"].fillna(raw_test_dataset["Departure Delay in Minutes"], inplace=True)
```

```python
raw_train_dataset.isnull().sum()
```

```python
raw_test_dataset.isnull().sum()
```

```python
combined_dataset = pd.concat([raw_train_dataset, raw_test_dataset], axis=0)
```

```python
combined_dataset.shape
```

```python
class_map = {'Business': 2, 'Eco Plus': 1, 'Eco': 0}
combined_dataset['Class'] = combined_dataset['Class'].map(class_map)
```

```python
combined_dataset['satisfaction'] = combined_dataset['satisfaction'].map({'neutral or dissatisfied': 0, 'satisfied': 1})
combined_dataset['Gender'] = combined_dataset['Gender'].map({'Male': 0, 'Female': 1})
combined_dataset['Customer Type'] = combined_dataset['Customer Type'].map({'disloyal Customer': 0, 'Loyal Customer': 1})
combined_dataset['Type of Travel'] = combined_dataset['Type of Travel'].map({'Personal Travel': 0, 'Business travel': 1})
```

```python
# Create a stacked bar chart of satisfaction vs. Inflight wifi service and Online boarding
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Satisfaction vs. Inflight wifi service
pd.crosstab(combined_dataset['Inflight wifi service'], combined_dataset['satisfaction']).plot(
    kind='bar', stacked=True, ax=axes[0], color=['#d62728', '#2ca02c']
)
axes[0].set_title('Satisfaction vs. Inflight Wifi Service')
axes[0].set_xlabel('Inflight Wifi Service Rating')
axes[0].set_ylabel('Count')
axes[0].legend(title='Satisfaction')
axes[0].tick_params(axis='x', rotation=0)

# Plot 2: Satisfaction vs. Online boarding
pd.crosstab(combined_dataset['Online boarding'], combined_dataset['satisfaction']).plot(
    kind='bar', stacked=True, ax=axes[1], color=['#d62728', '#2ca02c']
)
axes[1].set_title('Satisfaction vs. Online Boarding')
axes[1].set_xlabel('Online Boarding Rating')
axes[1].set_ylabel('Count')
axes[1].legend(title='Satisfaction')
axes[1].tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.show()
```

```python
# Create a countplot of satisfaction segmented by Class
plt.figure(figsize=(10, 6))
sns.countplot(data=combined_dataset, x='Class', hue='satisfaction', palette=['#d62728', '#2ca02c'])
plt.title('The Class Divide: Satisfaction by Class')
plt.xlabel('Class')
plt.ylabel('Count')
plt.legend(title='Satisfaction')
plt.tight_layout()
plt.show()
```

```python
# Create a scatter plot of Departure Delay vs. Arrival Delay
plt.figure(figsize=(10, 6))
plt.scatter(combined_dataset['Departure Delay in Minutes'], combined_dataset['Arrival Delay in Minutes'], alpha=0.5)
plt.xlabel('Departure Delay in Minutes')
plt.ylabel('Arrival Delay in Minutes')
plt.title('The Delay Distribution: Departure Delay vs. Arrival Delay')
plt.tight_layout()
plt.show()
```

```python
combined_dataset['Total_Delay'] = combined_dataset['Departure Delay in Minutes'] + combined_dataset['Arrival Delay in Minutes']
combined_dataset = combined_dataset.drop(['Departure Delay in Minutes', 'Arrival Delay in Minutes'], axis=1)
```

```python
# Select only numerical columns for correlation analysis
numerical_cols = combined_dataset.select_dtypes(include=[np.number]).columns

# Calculate correlation matrix
correlation_matrix = combined_dataset[numerical_cols].corr()

# Create heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0, 
            square=True, linewidths=0.5, cbar_kws={"shrink": 0.8})
plt.title('Correlation Heatmap of Numerical Features')
plt.tight_layout()
plt.show()
```
